```python
disease_categories = {
    1: "digestive system diseases",
    2: "cardiovascular diseases",
    3: "neoplasms",
    4: "nervous system diseases",
    5: "general pathological conditions"
}
```

```python
#check class ditribution
df['lables'].value_counts()

#apparently there isn't much class imbalance present here
```

```python
#lets find abstract token distribution
```

```python
def convert_to_num_words(abstract):
    return len(abstract.split())
```

```python
df['n_words'] = df['text'].apply(convert_to_num_words)
```

```python
#plot the distribution of the num of words
import matplotlib.pyplot as plt

# Example list of values
data = df['n_words'].values
# Plotting the histogram
plt.hist(data, bins=range(min(data), max(data) + 1), edgecolor='black', align='left')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of number of words in each abstract')
plt.show()
```

```python
#minum and maximum words in abstract
max(data) , min(data)
```

```python
#plot the distribution of the num of words for each class seperately
import matplotlib.pyplot as plt

# Example list of values
data = df['n_words'].values
# Plotting the histogram
plt.hist(data, bins=range(min(data), max(data) + 1), edgecolor='black', align='left')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of number of words in each abstract')
plt.show()
```

```python
#plot the distribution of the num of words for each class seperately
import matplotlib.pyplot as plt


for i in range(1,6):
# Example list of values
    data = df[df['lables']==i]['n_words'].values
    # Plotting the histogram
    plt.hist(data, bins=range(min(data), max(data) + 1), edgecolor='black', align='left')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of number of words in disease {disease_categories[i]} abstracts')
    plt.show()
```
